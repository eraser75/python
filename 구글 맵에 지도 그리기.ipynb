```python
import folium

m = folium.Map(
    location=[37.4178683,127.1253789],
    zoom_start=16
)

##기준점
folium.vector_layers.Circle(
  [37.418607, 127.126947],
  radius=500, ## 프레시 365 기준 500M
  color='#ffffgg',
  fill_color='#fffggg',
  popup='프레시365 라커'
).add_to(m)

folium.Marker(
  location=[37.418607, 127.126947],
  popup='프레시365 라커',
  icon=folium.Icon(color='red',icon='ok')
).add_to(m)

## 군집화
## 미트박스 회원
member = [
  {"store":"네이처키친","loc":[37.4172869,127.1262279]}
, {"store":"커리1978","loc":[37.417904, 127.124782]}
, {"store":"가온스시","loc":[37.417407, 127.126804]}
, {"store":"봉평메밀일번지","loc":[37.418830, 127.135949]}
, {"store":"더두부","loc":[37.418040, 127.126410]}
, {"store":"양정인순대국","loc":[37.417314, 127.125754]}
, {"store":"봄봄막국수","loc":[37.417393, 127.124871]}
, {"store":"미다원성남시청점","loc":[37.418142, 127.125748]}
, {"store":"라라멘","loc":[37.416894, 127.125645]}
]

## 미트박스 비회원
Non_members = [
  {"store":"놀부부대찌개&철판구이","loc":[37.418317, 127.125741]}
, {"store":"더쉐프퓨전철판요리","loc":[37.418369, 127.125272]}    
, {"store":"포츈차이니즈레스토랑","loc":[37.417358, 127.126878]}
, {"store":"덕발장","loc":[37.416871, 127.127513]}
, {"store":"할머니김밥","loc":[37.415296, 127.124560]}    
, {"store":"인사동토속음식점","loc":[37.419020, 127.130994]}    
, {"store":"고궁","loc":[37.419439, 127.131189]}        
, {"store":"자담치킨","loc":[37.419594, 127.131173]}            
]

from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(m)
marker_cluster2 = MarkerCluster().add_to(m)

for i in range(len(member)):
  folium.Marker(
    location=member[i]['loc'],
    popup=member[i]['store'],
    icon=folium.Icon(color='lightblue',icon='ok'),
  ).add_to(marker_cluster)

for i in range(len(Non_members)):
  folium.Marker(
    location=Non_members[i]['loc'],
    popup=Non_members[i]['store'],
    icon=folium.Icon(color='darkblue',icon='ok'),
  ).add_to(marker_cluster2)

''' 미박 본사
folium.Marker(
  location=[37.417875, 127.127494],
  popup='미트박스 본사',
  icon=folium.Icon(color='red',icon='ok')
).add_to(m)
'''
m.save('프레시365 라커.html')
```
